```python
# Create an empty one-hot encoded DataFrame
ohe_df = pd.DataFrame()

# Iterate over each property in the original DataFrame
for index, row in df.iterrows():
    fac = list(set(row['TopFacilities'].split()))

    # Create one-hot encoded columns for each facility
    for i in fac:
        if i not in ohe_df.columns:
            ohe_df[i] = 0  # Add the new facility column if it doesn't exist
        ohe_df.at[index, i] = 1  # Set the value to 1 for the corresponding property and facility

# Concatenate the one-hot encoded DataFrame with the original DataFrame
df = pd.concat([df, ohe_df], axis=1)

# Drop the original 'TopFacilities' column
df = df.drop('TopFacilities', axis=1)


```

```python
df
```

```python
df.set_index(df['PropertyName'],inplace=True)
```

```python
df.drop(columns=['PropertyName'],axis=1,inplace=True)
```

```python
df.fillna(0,inplace=True)
```

```python
df
```

Vectorization is done now we can calculate similarity score in vector space

```python
# cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity
# computing cosine similarity matrix
cos_sim=cosine_similarity(df)
# the cos_sim matrix contain the angular similarity to each vector to other
```

```python
cos_sim.shape
```

```python
property_name='Smartworld One DXP'
sim_score=cos_sim[df.index.get_loc(property_name)]
```

```python
sim_score # these all are the similarity score of that vector to other vector
```

```python
list(enumerate(sim_score)) # similarity with vector having those index in df
```

```python
def recomandation(sim_matrix,property_name):

  sim_score=list(enumerate(sim_matrix[df.index.get_loc(property_name)]))

  sorted_score=sorted(sim_score,key=lambda x:x[1],reverse=True)


  top_indices=[i[0] for i in sorted_score[1:6]]
  top_score=[i[1] for i in sorted_score[1:6]]

  top_property=df.index[top_indices].tolist()

  rec=pd.DataFrame({
      'property_suggestion':top_property,
      'score':top_score
  })

  return rec
```

```python
recomandation(cos_sim,property_name)
```

**Exporting cosine_similarity_matrix**

```python
import pickle
with open('facility.pkl', 'wb') as file:
    pickle.dump(cos_sim, file)
```
