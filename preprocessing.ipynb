```python
# converting into lower case
def lower(struct):
  con={}
  for loc,dist in struct.items():
    loc=loc.lower()
    dist=dist.lower()
    con[loc]=dist
  return con
new_df['LocationAdvantages']=new_df['LocationAdvantages'].apply(lower)
```

```python
new_df['LocationAdvantages'][0]
```

```python
# convert into km
def convert_km(struct):

  for loc,dist in struct.items():
    try:
      value,unit=dist.split()
      if unit=='meter' or unit=='m':
        value=round(float(value)/1000,3)
        struct[loc]=f'{value} km'
    except:
      pass



new_df['LocationAdvantages'].apply(convert_km)

new_df['LocationAdvantages'][0]
```

```python
loc_data=new_df[['LocationAdvantages','PropertyName']]
```

```python
loc_data
```

```python
def make_column(df):
  locations={}
  for index,rows in df.iterrows():
    dist={}
    for location,distance in rows['LocationAdvantages'].items():
      dist[location]=distance

    locations[index]=dist
  new=pd.DataFrame(locations).transpose()
  return new


new=make_column(loc_data)
```
