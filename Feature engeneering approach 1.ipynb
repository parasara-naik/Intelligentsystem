```python
import re
def get_superbuilt(text):
    match=re.search('super built up area[:]? (\d*\.?\d+) {0,2}?\((\d*\.?\d+) sq.m.\)',text)
    if match:
        return float(match.group(2))# return the 2nd capturing group
    else:
        return np.nan
        
```

```python
df['super_bulituparea(sqm)']=df['areaWithType'].astype(str).apply(get_superbuilt)
```

```python
df['copy']=df['areaWithType']
```

```python
def remove(text):
    formated_text=re.sub(r'super built up area[:]? (\d*\.?\d+) {0,2}?\((\d*\.?\d+) sq.m.\)','',text)
    return formated_text
```

```python
df['copy']=df['copy'].astype(str).apply(remove)
```

```python
def get_builtarea(text):
    match=re.search('built up area[:]? (\d*\.?\d+) (sq\.ft\.)?(sq\.yards)? {0,2}?\((\d*\.?\d+) sq.m.\)',text)
    if match:
        return float(match.group(4))
    else:
        return np.nan
```

```python
df['built_uparea(sq-m)']=df['copy'].astype(str).apply(get_builtarea)
```

```python
def get_carpetarea(text):
    match=re.search('carpet area[:]? (\d*\.?\d+) (sq\.ft\.)?(sq\.yards)? {0,2}?\((\d*\.?\d+) sq.m.\)',text)
    if match:
        return float(match.group(4))
    else:
        return np.nan
```

```python
df['carpet_area(sqm)']=df['copy'].astype(str).apply(get_carpetarea)
```

```python
def remove2(text):
    formatted_text=re.sub('carpet area[:]? (\d*\.?\d+) (sq\.ft\.)?(sq\.yards)? {0,2}?\((\d*\.?\d+) sq.m.\)','',text)
    return formatted_text
```

```python
def remove3(text):
    formatted_text=re.sub('built up area[:]? (\d*\.?\d+) (sq\.ft\.)?(sq\.yards)? {0,2}?\((\d*\.?\d+) sq.m.\)','',text)
    return formatted_text
```

```python
df['copy']=df['copy'].astype(str).apply(remove3)
```

```python
df['copy']=df['copy'].astype(str).apply(remove2)
```

```python
def get_plotarea(text):
    match=re.search('plot area (\d*\.?\d+)\((\d*\.?\d+) sq\.m\.\)',text)
    if match:
        return float(match.group(2))
    else:
        return np.nan
```

```python
def remove4(text):
    formatted_text=re.sub('plot area (\d*\.?\d+)\((\d*\.?\d+) sq\.m\.\)','',text)
    return formatted_text
```

```python
df['built_uparea(sq-m)'].isnull().sum()
```

```python
'''In terms of similarity to "plot area," the closest match would be the buiit up area.
the total area you own in a city or town is called plot area.'''
df['built_uparea(sq-m)'] = df.apply(lambda row: get_plotarea(row['copy']) if row['copy'] != '' else row['built_uparea(sq-m)'], axis=1)
```

```python
df['copy']=df['copy'].astype(str).apply(remove4)
```

```python
def get(text):
    match=re.match('carpet area: (\d*\.?\d+)',text)
    if match:
        return round((float(match.group(1))*0.092903),2)
    else:
        return np.nan
```

```python
df['carpet_area(sqm)']=df.apply(lambda x:get(x['copy']) if x['copy']!='' else x['carpet_area(sqm)'],axis=1)
```

```python
def remove5(text):
    formatted_text=re.sub('carpet area: (\d*\.?\d+)','',text)
    return formatted_text
```

```python
def categorize_age_possession(value):
    if pd.isna(value):
        return "Undefined"
    if "0 to 1 Year Old" in value or "Within 6 months" in value or "Within 3 months" in value:
        return "New Property"
    if "1 to 5 Year Old" in value:
        return "Relatively New"
    if "5 to 10 Year Old" in value:
        return "Moderately Old"
    if "10+ Year Old" in value:
        return "Old Property"
    if "Under Construction" in value or "By" in value:
        return "Under Construction"
    try:
        # For entries like 'May 2024'
        int(value.split(" ")[-1])
        return "Under Construction"
    except:
        return "Undefined"
```

```python
df['agePossession'] = df['agePossession'].apply(categorize_age_possession)
```
